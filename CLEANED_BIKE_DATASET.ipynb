About Dataset:
This dataset consist of bike-sales reords accross countries from the year 2021. The dataset is currently unclean; concsistsing of missing values, incocnsistent data types, duplicated values.
The dataset has to be cleaned before any kind of analysis can be done.

Data source: Kaggle
For more info about the datset, please visit [here](https://www.kaggle.com/datasets/ratnarohith/uncleaned-bike-sales-data)



Columns Description:
*   Sales_Order : This column contains contains unique sales order number.
*   Date : This column contains the purchase date.
*   Day : This column contains the day number of the purchase.
*   Month : This column contains the month of the purchase.
*   Year : This column contains the year of the purchase
*   Customer_Age : This column contains age of the customer.
*   Age_Group : This column contains the age group of the customer. They are grouped as ('Adults (35-64)', 'Young Adults (25-34)', 'Youth (<25)).
*   Customer_Gender: This column contains the gender of the customer.
*   Country : This column contains the country where the purchase was made.
*   State : This column contains the state where the purchase was made.
*   Product_Category :This column contains the name of the product purchased(bike).
*   Sub_Category : This column contains the sub-category of the product purchased( what kind of bike was purchased).
*   Product_Description : This column contains the full description of the product purchased.
*   Order_Quantity : This column contains the amount of procuct(bike) purchased.
*   Unit_Cost : This column contains the cost per unit of the product(bike).
*   Unit_Price : This column contains the selling price per unit of the product.
*   Profit : This column contains the profit made from each other.
*   Cost : This column contains the cost incurred for the order (Order_Quantity × Unit_Cost).
* Revenue – This column contains the total revenue generated from the order (Order_Quantity × Unit_Price).









Issues with the data:
*   Missing values
*   Inconsistent data type
*   Duplicate rows

DATA CLEANING

*   Handling data columns names

```python
data.head()
```

```python
# Viewing the column names
data.columns
```

```python
# Stripping the column names of extra spaces
data.columns = data.columns.str.strip()


# Renaming column names
data.rename(columns={'Sales_Order #' : 'Sales_Order_No'}, inplace=True)
```

```python
# View the updated column names
data.columns
```

```python
# The dataset already has a 'date' column, so the 'Day', 'Month', 'Year'columns are unnecessary
# Dropping unnecessary columns
data.drop(['Day', 'Month', 'Year'], axis=1, inplace=True)
```

* Handling extra spaces in string-values

```python
# Removing extra spaces from string-values in the dataset
for col in data.select_dtypes(include='object').columns:
    data[col] = data[col].str.strip()
```

* Handling duplicates

```python
# Total duplicated rows
data.duplicated().sum()
```

```python
# Identifying the duplicted row
data[data.duplicated()]
```

```python
# Deleting the duplicated row
data.drop_duplicates(inplace=True)
```

```python
# After dropping a row, index tends to be disorganized.
# Reseting the index.
data.reset_index(drop= True, inplace=True)
```

* Handling missing values

```python
# Checking for missing values in each column
data.isnull().sum()
```

```python
# Filtering the rows with their missing values
data[data.isnull().any(axis=1)]
```

```python
# Filling missing value in index 14

customer_age = data.loc[14, 'Customer_Age']

if customer_age < 25:
    data.loc[14, 'Age_Group'] = 'Youth (<25)'
elif 25 <= customer_age <= 34:
   data.loc[14, 'Age_Group'] = 'Young Adults (25-34)'
elif 35 <= customer_age <= 64:
   data.loc[14, 'Age_Group'] = 'Adults (35-64)'
```

```python
# Filling missing value in index 20


# Compared bikes with simiilar Unit_Cost, Unit_Price, Profit, Cost, Revenue to fill Product_Descriptio
# The Unit_cost of the missing column was 1252, and that price matches to 'Mountain-200 Black'
# Now to specify the type of 'Mountain-200 Black', I filled the column with the most frequent 'Mountain-200 Black' which is :'Mountain-200 Black, 46'


from statistics import mode
most_frequent_model = mode(data.loc[data['Product_Description'].str.contains('Mountain-200 Black', na=False), 'Product_Description'])
print(most_frequent_model)

# Updating the empty column with the most frequent model
data.loc[20, 'Product_Description'] = most_frequent_model    
```

```python
# Filling missing value in index 21
# Compared bikes with simiilar Unit_Cost, and noticed that an Order_Quantity of 1 has a unit_Cost of: 295, Unit_Price of : 540, profit of: 245

rows_with_unit_cost_295 = data[data["Unit_Cost"] == 295]
print(rows_with_unit_cost_295)

data.loc[21, 'Order_Quantity'] = 1              # Updating the missing value to 1
```

* Handling incorrect values/ inconsistencies

```python
# Updating the duplicated 'Sales_Order_No' of index 1
data.loc[1, 'Sales_Order_No'] = 261696
```

```python
# Updating incorrect values in index 4('Unit_Cost' and 'Cost' column)
# 'Profit' = 'Revenue' - 'Cost'
# Since  'Cost' is missing, therefore, 'Cost' = 'Revenue' - 'Profit

data.loc[4, 'Cost'] = data.loc[4, 'Revenue'] - data.loc[4, 'Profit']        



# 'Cost' = 'Unit_Cost' * ' Order_Quantity'
# Therefore, 'Unit_Cost' will be = 'Cost' / 'Order_Quantity'
data.loc[4, 'Unit_Cost'] = data.loc[4, 'Cost'] / data.loc[4, 'Order_Quantity']      
```

```python
# Updating incorrect values in index 7('Unit_Cost' and 'Revenue' column)
# 'Profit' = 'Revenue' - 'Cost' Therefore, 'Revenue' will be = 'Profit' + 'Cost'
data.loc[7, 'Revenue'] = data.loc[7, 'Profit'] + data.loc[7, 'Cost']


# 'Revenue' = 'Unit Price' * 'Order Quantity' Therefore, 'Unit Price' = ''Revenue' / 'Order Quantity'
data.loc[7, 'Unit_Price'] = data.loc[7, 'Revenue'] / data.loc[7, 'Order_Quantity']
```

```python
# Updating incorrect values in index 21('Cost' and 'Revenue' column)

data.loc[21, 'Cost'] = data.loc[21, 'Unit_Cost'] * data.loc[21, 'Order_Quantity']

data.loc[21, 'Revenue'] = data.loc[21, 'Profit'] + data.loc[21, 'Cost']
```
